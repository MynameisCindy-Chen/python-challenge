```python
# Import csv file
import csv

# set the csv file path
budget_data = "C:/Users/Ivan/Downloads/Starter_Code/Instructions/PyBank/Resources/budget_data.csv"

# Initialize line_num variable
line_num = 0

# open the csv file
with open(budget_data, "r") as csvfile:
    
    # print the datatype of the file object
    print(type(csvfile))
    
    # pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")
    # Print the datatype of the csvreader
    print(type(csvreader))
    print()
    
    # Go to the next row from the start of the file
    header = next(csvreader)
    # list line starts with header
    line_num += 1
    # print the header
    print(f"{header} < ----- HEADER")
    
    # Initialize mertic variables
    total_months = 0
    total_profits_losses = 0
    avg_of_profits_losses = 0
    
    maximum_month = {}
    maximum_profits = 0
    minimum_month = {}
    minimum_profits = 0
      
    # Create a list to include the changes in profit/losses by date
    net_profits_losses = []
    
    # Skip first line of data to assign first line value
    skip_header = next(csvreader)
    first_month = skip_header[1]
    # Calculate first line of data 
    total_months = total_months + 1
    total_profits_losses = total_profits_losses + int(first_month)
    
    # read each row of data after the header
    for row in csvreader:  
        
        # print each row
        print(skip_header)
        print(row)
        
        # Calculate total number of months
        total_months = total_months + 1
        # Calculate the net total amount of profit/losses
        total_profits_losses += int(row[1])
        
        # Calcute the the change in profit/losses date by date
        difference = int(row[1]) - int(first_month)
        first_month = int(row[1])
        # Add the difference to the list
        net_profits_losses.append(difference)
        
        # To get the greatest increase in profits and greatest decrease in losses (date and amount)
        if maximum_profits == 0:
            maximum_profits = int(row[1])
            maximum_month = row
        elif int(row[1]) > maximum_profits:
            maximum_profits = int(row[1])
            maximum_month = row
        elif int(row[1]) < minimum_profits:
            minimum_profits = int(row[1])
            minimum_month = row
        
# Caculate the average changes in profit/losses
net_profits_losses = round(sum(net_profits_losses) / total_months,2)

# Output the result to a text file
file_to_output = "budget_analysis.txt"

# Open the text file and write the result
with open(file_to_output, "w") as txt_file:
    txt_file.write(f"Fianacial Analysis\n")
    txt_file.write(f"----------------------------\n")
    txt_file.write(f"Total Months: {total_months}\n")
    txt_file.write(f"Total: ${total_profits_losses}\n")
    txt_file.write(f"Average Change: ${net_profits_losses}\n")
    txt_file.write(f"Greatest Increase in Profits: {maximum_month[0]} (${maximum_month[1]}) \n")
    txt_file.write(f"Greatest Decrease in Profits: {minimum_month[0]} (${minimum_month[1]}) \n")
    

print()
print(maximum_month)
print("Financial Analysis")
print("--------------------------------")
print(f"Total Months: {total_months}")
print(f"Total: ${total_profits_losses}")
print(f"Average Change: ${net_profits_losses}")
print(f"Greatest Increase in Profits: {maximum_month[0]} (${maximum_month[1]}) ")
print(f"Greatest Decrease in Profits: {minimum_month[0]} (${minimum_month[1]}) ")
```
